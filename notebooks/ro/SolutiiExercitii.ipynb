```python
import math
class Model:
  def __init__ (self, valoare, radical):
    self.valoare = valoare
    self.radical = radical

  @staticmethod
  def calc_radical(val):
    return math.sqrt(val)

  @classmethod
  def init(cls, val):
    return Model(val, cls.calc_radical(val))

  def __str__(self):
    return ("Valoare: %.2f, radical: %.2f" %(self.valoare, self.radical))


O = Model(9, 3)
print(O)
print(Model.calc_radical(25))
O2 = Model.init(81)
print(O2)
```

5) Scrieți o aplicație care definește o clasă de verificare a unei chei de autentificare.
Cheia de autentificare este de tipul: `XXXXX-XXXXX-XXXXX-XXXXX`, unde X reprezintă un caracter ce poate fi cifră sau literă. Cheia are exact 4 grupuri de caractere a câte 5 caractere fiecare, separate prin caracterul ‘-‘. De asemenea, numărul de cifre trebuie să fie mai mare decât numărul de litere, iar numărul de litere nu poate să fie 0. În cazul în care nu este îndeplinită cel puțin o condiție din cele menționate anterior, se aruncă o excepție proprie cu mesajul: ”Cheie de autentificare incorectă!”. 
Toate verificările se fac în momentul instanțierii unui nou obiect din clasa definită.

```python
class CheieError(Exception):
  pass

class Cheie:
  def numar_litere(self,S):
    numar_lit = len([c for c in S if c.isalpha()])
    if numar_lit != 0:
      return 1
    return 0
  def __init__(self, S):
    if len(S.split('-')) != 4:
      raise CheieError("Numarul de grupuri e incorect")
    elif self.numar_litere(S) == 0:
      raise CheieError("Numarul de litere e 0")
    elif not self.numar_caractere_grup(S):
      raise CheieError("Numarul de caractere din grup e incorect")
    elif not self.numar_cifre_numar_litere(S):
      raise CheieError("Numarul de cifre e mai mare decat numarul de litere")
    else:
      self.cheie = S
      print("Cheie inițializată cu succes")

  
  
  def numar_caractere_grup(self, S):
    for grup in S.split('-'):
      if len(grup)!=5:
        return 0
    return 1

  def numar_cifre_numar_litere(self,S):
    nc = 0 # număr caractere
    nn = 0 # număr cifre
    for c in S:
      if c.isalpha():
        nc+=1
      elif c.isnumeric():
        nn+=1
    if nc > nn:
      return 1
    else:
      return 0
```

```python
from tempfile import TemporaryFile
with TemporaryFile('w+') as f:
  # primii doi termeni
  n1, n2 = 0, 1
  count = 0

  while count < 10:
      f.write(str(n1)+' ')
      val_fib = n1 + n2
      # actualizăm valorile n1 și n2
      n1 = n2
      n2 = val_fib
      count += 1
  f.seek(0)
  data = ' '.join(f.read().split()[::-1])
  print(data)
```
